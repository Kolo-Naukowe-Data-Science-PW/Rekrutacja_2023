```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dane.csv")
df.head(10)
```

```python
df.tail()
```

```python
passengers = df.shape[0]
df.shape
```

W sumie było 8693 pasażerów.

```python
df.describe()
```

```python
transportedTrue  = len(df.loc[(df["Transported"] == True)])
transportedFalse = len(df.loc[(df["Transported"] == False)])
print("Sukces podróży: ", transportedTrue  / passengers * 100, "%")
print("Porażka podróży:", transportedFalse / passengers * 100, "%")
```

```python
categories = ["Przetransportowani", "Nieprzetransportowani"]
number = [transportedTrue, transportedFalse]

plt.bar(categories, number, color=['green', 'red'])
plt.ylabel("Liczba Pasażerów")
plt.title("Sukces podróży")
plt.show()
```

Tylko około połowy pasażerów zostało przestransportowanych z sukcesem. Różnica jest prawie niezauważalna.

```python
vips = vipTransportedTrue  = len(df.loc[(df["VIP"] == True)])
vipTransportedTrue  = len(df.loc[(df["Transported"] == True) & (df["VIP"] == True)])
vipTransportedFalse = len(df.loc[(df["Transported"] == False) & (df["VIP"] == True)])
print("Procent vip wśród pasażerów:", vips  / passengers * 100, "%")
print("Przetransportowani vip:     ", vipTransportedTrue  / vips * 100, "%")
print("Nie przetransportowani vip: ", vipTransportedFalse / vips * 100, "%")
```

```python
categories = ["Przetransportowani vip", "Nieprzetransportowani vip"]
number = [vipTransportedTrue, vipTransportedFalse]

plt.bar(categories, number, color=['green', 'red'])
plt.xlabel("Status Przetransportowania")
plt.ylabel("Liczba Pasażerów")
plt.show()
```

Tylko 38% vipów zostało skutecznie przetransportowanych. 
Wniosek: posiadanie statusu vip nie zwiększa w istotny sposób szansy na zostanie przetransportowanym.
Liczba osób vip jest bardzo mała w stosunku do ilości pasażerów, przez co informacja ta może być niemiarodajna.

```python
cryoSleep = len(df.loc[(df["CryoSleep"] == True)])
cryoSleepAndTransportedTrue  = len(df.loc[(df["Transported"] == True) & (df["CryoSleep"] == True)])
cryoSleepAndTransportedFalse = len(df.loc[(df["Transported"] == False) & (df["CryoSleep"] == True)])
print("Procent osób w kriośnie wśród pasażerów:", cryoSleep  / passengers * 100, "%")
print("Przetransportowane osoby w kriośnie:    ", cryoSleepAndTransportedTrue  / cryoSleep * 100, "%")
print("Nie przetransportowane osoby w kriośnie:", cryoSleepAndTransportedFalse / cryoSleep * 100, "%")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(["Przetransportowani", "Nieprzetransportowani"], [cryoSleepAndTransportedTrue, cryoSleepAndTransportedFalse])
plt.ylabel("Liczba pasażerów w kriośnie")
plt.title("Pasażerowie w kriośnie")
plt.show()
```

Większość osób w kriośnie zostało z powodzeniem przetransportowanych. 
Wniosek: bycie w kriośnie znacząco zwiększa prawdopodobieństwo bycia przetransportowanym.
Ilość osób w kriośnie w stosunku do liczby pasażerów jest reprezentatywna.

```python
df["AdditionalCosts"] = df["RoomService"] + df["FoodCourt"] + df["ShoppingMall"] + df["Spa"] + df["VRDeck"]
df = df.drop(["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"], axis=1)
df.head()
```

```python
withAdditionalCost = len(df.loc[(df["AdditionalCosts"] != 0.0)])
withAdditionalCostAndTransported = len(df.loc[(df["Transported"] == True) & (df["AdditionalCosts"] != 0.0)])
withHugeAdditionalCostAndTransported = len(df.loc[(df["Transported"] == True) & (df["AdditionalCosts"] > 10000)])
withoutAdditionalCostAndTransported = len(df.loc[(df["Transported"] == True) & (df["AdditionalCosts"] == 0.0)])

print("Pasażerowie z dodatkowymi kosztami: ", withAdditionalCost / passengers * 100, "%")
print("Pasażerowie z dodatkowymi kosztami i przetransportowani (spośród tych z dodatkowymi kosztami): ", withAdditionalCostAndTransported / withAdditionalCost * 100, "%")
print("Pasażerowie z dużymi (> 10 000) dodatkowymi kosztami i przetransportowani: ", withHugeAdditionalCostAndTransported / withAdditionalCostAndTransported * 100, "%")
print("Pasażerowie z dodatkowymi kosztami i przetransportowani (spośród wszystkich przetransportowanych): ", withAdditionalCostAndTransported / transportedTrue * 100, "%")
print("Pasażerowie bez dodatkowych kosztów i przetransportowani (spośród wszystkich przetransportowanych): ", withoutAdditionalCostAndTransported / transportedTrue * 100, "%")
```

Większość pasażerów którzy zostali pomyślnie przetransportowani nie ponosiła dodatkowych kosztów w trakcie podróży (58.13%).
Tylko 41.87% pasażerów przetransportowanych pomyślnie poniosło dodatkowe koszty.
Tylko co trzeci pasażer, który poniósł dodatkowe koszty został pomyślnie przetransportowany. 
Wniosek: ponoszenie dodatkowych kosztów nie ma pozytywnego wpływu na pomyślność przetransportowania.

```python
route_counts = df.groupby(["HomePlanet", "Destination", "Transported"]).size().unstack(fill_value=0)

route_counts.plot(kind="bar", stacked=True)
plt.xlabel("Trasa")
plt.ylabel("Liczba pasażerów")
plt.title("Transport wg. trasy")
plt.xticks(rotation=90) 
plt.legend(title="Status transportu", labels=["Nieprzetransportowani", "Przetransportowani"])
plt.ylim(0, 3150)
plt.show()
```

```python
df["CabinLetter"] = df["Cabin"].str[0]

plt.figure(figsize=(8, 6))
ax = df.groupby(["CabinLetter", "Transported"]).size().unstack(fill_value=0).plot(kind="bar", stacked=True)
plt.xlabel("Kabina")
plt.ylabel("Liczba Pasażerów")
plt.title("Zależność przetransportowania od rodzaju kabiny")
plt.xticks(rotation=0)
plt.legend(title="Status Transportu", labels=["Nieprzetransportowani", "Przetransportowani"])
plt.show()
```

Najwięcej pasażerów zazwyczaj wybierało kabiny F i G.
Największą szanse na udany transport mieli pasażerowie z kabin B, C i G (więcej niż 50%).
*kabina T była wybierana rzadko, stąd jej zerowy wygląd na wykresie

```python
bins = [0, 17, 30, 40, 50, float('inf')]
labels = ['0-17', '18-30', '31-40', '41-50', '51+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

age_group_counts = df.groupby(["AgeGroup", "Transported"]).size().unstack(fill_value=0)

plt.figure(figsize=(8, 6))
age_group_counts.plot(kind="bar", stacked=True)
plt.xlabel("Grupa wiekowa")
plt.ylabel("Liczba pasażerów")
plt.title("Transport wg. grupy wiekowej")
plt.xticks(rotation=0)
plt.legend(title="Status transportu", labels=["Nieprzetransportowani", "Przetransportowani"])
plt.show()
```

Pasażerowie w wieku 18-30 są najbliżej średniej tj. 50% przetransportowanych.
Największe prawdopodobieństwo na udany transport mają osoby w wieku 0-17 i 51+ (więcej niż 50%).
